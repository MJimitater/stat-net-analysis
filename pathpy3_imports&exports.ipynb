```python
import pathpy as pp
import pandas as pd
```

```python
n = pp.Network()
n.add_edge('a','b',weight=2)
n.add_edge('a','c',type='business partners')
print(n)
n
```

## Exporting network to CSV

Creates CSV file of network, containing source nodes and target nodes and their edge UID.

With parameter `sep`, we can define the symbol of separation.

```python
pp.io.write_csv(n, 'network.csv', sep=';')
```

## Exporting CSV to network 

```python
n = pp.io.read_csv('network.csv', sep=';')
print(n)
```

## Exporting network to Pandas Dataframe

Edges that don't have a certain attribute, are assigned with `NaN`. 

```python
dataframe = pp.io.to_dataframe(n)
dataframe
```

## Exporting Pandas Dataframe to network

Name the column for the source node `v`.

Name the column for the target node `w`.

If no column `uid` is provided, then new edge UIDs will be created.

```python
#Create empty DataFrame:
dataframe = pd.DataFrame(columns=['v', 'w', 'uid'])

#Fill rows of DataFrame:
dataframe.loc[0] = ['a', 'b', 'ugotpranked']
dataframe.loc[1] = ['b', 'c', 'ugothacked']

dataframe
```

```python
n = pp.io.from_dataframe(dataframe)
print(n)
```

```python
print(n.nodes.uids)
print(n.edges.uids)
```

## Exporting SQLite Database to network


`pp.io.read_sql(filename='database.db', table='highschool', directed=True)` reads an SQLite database into a network.

The elements in the column named `v` will be read as the source nodes.

The elements in the column named `w` will be read as the target nodes.

Any other elements in another column will be read as an edge attribute.


```python
gentoo = pp.io.read_sql(filename='networks.db', table='gentoo')

gentoo.plot()
```

## Exporting network to SQLite Database


`pp.io.write_sql(filename='database.db', table='highschool', if_exists='replace')` exports a network into a SQLite database.
